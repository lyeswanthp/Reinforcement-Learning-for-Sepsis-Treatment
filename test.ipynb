```python
import pandas as pd

# Load the data
train_df = pd.read_csv('data/processed/train_trajectories.csv')
val_df = pd.read_csv('data/processed/val_trajectories.csv')
test_df = pd.read_csv('data/processed/test_trajectories.csv')

# Get unique patient IDs
train_ids = set(train_df['stay_id'].unique())
val_ids = set(val_df['stay_id'].unique())
test_ids = set(test_df['stay_id'].unique())

# Check overlaps
train_val_overlap = train_ids & val_ids
train_test_overlap = train_ids & test_ids
val_test_overlap = val_ids & test_ids

print("="*50)
print("PATIENT OVERLAP CHECK")
print("="*50)
print(f"Train patients: {len(train_ids)}")
print(f"Val patients:   {len(val_ids)}")
print(f"Test patients:  {len(test_ids)}")
print()
print(f"Train-Val overlap:  {len(train_val_overlap)}")
print(f"Train-Test overlap: {len(train_test_overlap)}")
print(f"Val-Test overlap:   {len(val_test_overlap)}")
print()

if len(train_test_overlap) == 0 and len(train_val_overlap) == 0:
    print("✅ NO DATA LEAKAGE - Splits are clean")
else:
    print("❌ DATA LEAKAGE DETECTED - Need to re-split by stay_id")
```
